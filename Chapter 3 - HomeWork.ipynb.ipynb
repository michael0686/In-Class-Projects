```python
#twoDimensionalListAndNumpyArray.py
import numpy as np

two_dim_list = [[1, 2, 3, 4], [2, 3, 4, 5]]
print(two_dim_list)
len(two_dim_list)
#Here we create a two-dimensional list of length 2 which contains
#two seperate lists of length 4
```

```python
for i in range(len(two_dim_list)):
    print(two_dim_list[i])
```

```python
for i in range(len(two_dim_list)):
    print(two_dim_list[i])
    for j in range(len(two_dim_list[i])):
        print(two_dim_list[i][j])
#We can use a nested for loop to print each item from each list
#using the iterations as the subscript
```

```python
two_dim_array = np.array(two_dim_list)
print(two_dim_array)
#The numpy array function can coerce a multi-dimensional list
#to an array 
```

```python
for i in range(len(two_dim_array)):
    print(two_dim_array[i])
    for j in range(len(two_dim_array[i])):
        print(two_dim_array[i][j])
#Here we print each iterable value (j) for each iterable array (i)
#for a total of eight iterations over the dimensions of the 2D 
#array. First print two_dim_array[0] for i=1, next print 
#two_dim_array[0][0-3] for i=1 and j=1-4 then print 
#two_dim_array[1] for i=2, finally print two_dim_array[0][4-7] 
#for j=5-8.
```

```python
import numpy as np

array = np.zeros((3, 3))
print("Original Array:", array, sep = "\n")
#The zeros function in numpy creates an array of zeros with 
#dimensions specifed by its arguments. In this case 3x3.

array[0][2] = 9
array[1][0] = 7
array[2][2] = 3
print("Updated Array:", array, sep = "\n")
#We can use subscripting to change the values in an existing array.
#For example, array[0][2] = 9 in a 3x3 array of zeros will change
#the value in the first row and third column to 9.
```

```python
#zerosOnesAndLike.py  
import numpy as np  
list_of_lists = [[1,2,3], [4,5,6], [7,8,9]]  
array_of_arrays = np.array(list_of_lists)  
zeros_like_array = np.zeros_like(list_of_lists)  
ones_like_array = np.ones_like(list_of_lists)  
empty_like_array = np.empty_like(list_of_lists)  
 
print("list_of_lists:", list_of_lists, sep="\n")  
print("array_of_arrays:", array_of_arrays, sep="\n")  
print("zeros_like_array:", zeros_like_array, sep="\n")  
print("ones_like_array:", ones_like_array, sep="\n")  
print("empty_like_array:", empty_like_array, sep="\n")
#Numpy contains functions which can coerce arrays to contain
#different values based on the function called and the arguments.
#for example, if we call the np.oneslike function and use three
#arrays of length 3 as arguments, python will return a 3x3 array
#containing 1s.
```

```python
#numpyLogs.py
import numpy as np

vals = np.arange(0, 10001, 100)
ln_vals = np.log(vals)
log10_vals = np.log10(vals)
print(vals)
print(ln_vals)
print(log10_vals)
#Numpy has useful math functions like the log function. If we pass
#an array to the log function it will return an array of equal 
#length with each value reflecting the power that the euler 
#number must be raised to in order to return that value from
#the original array
```

```python
import numpy as np

pi = np.pi
e = np.e
lne = np.log(e)
infinity = np.inf
null_val = np.nan
print("pi =", pi)
print("e =", e)
print("ln(e) =", lne)
print("infinity =", infinity)
print("null_val =", null_val)
#Numpy as special values such pi=3.14, e=2.71, infinity, etc.
#for example, to call pi use np.pi
```

Plotting the Distribution of Income

```python
import pandas as pd
income_data = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])
#Use !pip install xlrd to read in an excel file that contains
#income data from the U.S. Census Bureau 
```

```python
income_data = income_data.rename(columns = {"Unnamed: 0": "Household Income"})
income_data
```

```python
income = income_data["Number"][1:-1].div(1000)
income.iloc[0] = 0
income.head()
#Data cleaning makes data easier to read. Here we create a series
#for household income that we've drawn from our income data.
#first we call the column of interest from the data, next we
#call the rows of interest (Drop the 1st and last row), finally
#we divide by 1000 to convert from thousands of households 
#to millions of households
```

```python
income_levels = list(range(0, 201, 5))
income_levels.append(250)
income_levels = [str(i) + "k" for i in income_levels]
income_levels.append("250k+")
print(income_levels)
#Create a list of the data that includes values from 0-200 in
#increments of 5. append 250 to the list. Convert each value to
#a string and concatenate 'k' to the string for thosands of 
#dollars. Lastly, append '250k+'.
```

```python
import pandas as pd
income_data = pd.DataFrame({"Income Level":np.array(income_levels),
                           "Households in Bracket":income})
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()
income_data = income_data.fillna(0).set_index("Income Level")
income_data.head()
```

```python
income_data["% in Bracket"] = income_data["Households in Bracket"].div(income_data["Households in Bracket"].sum()).mul(100)
income_data["Cumulative %"] = income_data["% in Bracket"].cumsum()
income_data.head()
```

```python
import matplotlib as plt
plt.rcParams.update({"font.size" : 16})
fig, ax = plt.subplots(figsize = (15, 7))
plot_df = income_data.dropna()
plot_df.plot.bar( 
                                 y=["% in Bracket","Cumulative %"],
                                 secondary_y = "Cumulative %", 
                                 fill=True, 
                                 alpha=.75, 
                                 width = .99,
    legend = True,
                                 ax=ax)

# set x labels to show income bins and add 250k+ bin
x_ticklabels = ["$" + str(x) for x in plot_df.index]

# mean income per capita is 67.5k, exactly halfway between 65k and 70k
ax.set_xticklabels(x_ticklabels)
index = list(plot_df.index).index("65k")
ax.axvline(index + .5, linewidth = 15, color = "C3", alpha = .5, label = "Mean")
ax.set_title("Mean: $67,500\nIncome Distribution in the United States")
```

```python
income_25k_bin = income_data.loc[::5]
income_25k_bin["Households in Bracket"] = income_25k_bin["Cumulative"]
income_25k_bin.loc["250k+"] = income_data.iloc[-1]
income_25k_bin.loc["250k+"]["Households in Bracket"] = income_25k_bin.loc[
    "250k+"]["Cumulative"] 
income_25k_bin["Households in Bracket"]= income_25k_bin[
    "Households in Bracket"].diff()
# income_25k_bin["% in Bracket"] = income_25k_bin["Households in Bracket"].div(income_25k_bin["Households in Bracket"].sum()).mul(100)
income_25k_bin.fillna(0, inplace = True)
income_25k_bin.head()
```

```python
import matplotlib as plt
plt.rcParams.update({"font.size" : 20})
plot_df = income_25k_bin[["% in Bracket", "Cumulative %"]].iloc[1:]
fig, ax = plt.subplots(figsize = (15, 10))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .99, 
                 alpha = .75,
                 color = ["C0","C1"],
                 ax = ax)

x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
ax.set_title("Income Distribution in the United States")
```
