Create a DataFrame from a Dictionary

```python
macro_dict = {"GDP": {},
             "Money": {},
             "Real GDP": {}}
macro_dict
```

```python
for key in macro_dict:
    for i in range(1990, 2018):
        macro_dict[key][i] = round(np.random.random() * 10000,2)
print(macro_dict)
```

```python
macro_df = pd.DataFrame(macro_dict)
macro_df.tail()
```

```python
macro_df["Price Level"] = macro_df["GDP"].div(macro_df["Real GDP"])
macro_df["Velocity"] = macro_df["GDP"].div(macro_df["Money"])
macro_df.head()
```

```python
macro_df.loc[1995:2000, ["Real GDP", "Money"]]
```

```python
macro_df.loc[[1995,1997,1999], ["Real GDP", "Money"]]
```

```python
macro_df.loc[[year for year in range(1990,2018, 4)], ["Real GDP", "Money"]]
```

```python
macro_df.loc[:, ["Real GDP"]]
#returns all years from the dataframe as the rows
```

```python
macro_df.loc[macro_df.index, macro_df.keys()]
#The index is the rows (years) and the keys are the columns (Variables).
```

```python
income_data = pd.read_excel(
    "https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", 
    header=[7])
#read_excel takes either a filepath or a URL as its first argument which allows python to find the excel file 
#that you want to read in.
```

```python

income_data
```

```python
households = income_data["Number"][1:-1].div(1000)
households.iloc[0] = 0
households
#'Number' is the column name for households in the income_data set. In the original data the first row and 
#last row had irrelevant information, so we call the rows from index 1 to index -1 to leave out these unneeded
#rows. Next we divide each value in the column by 1000 to make them smaller and easier to read. Moreover,
#the value at index[1] from the original data was the total number of households from the data, so we set
#this value equal to zero using its new index in the 'households' object which is index[0].
```

```python
income_levels = list(range(0,201,5))
income_levels.append(250)
income_levels = [str(value) + "k" for value in income_levels]
income_levels.append("250k+")
print(income_levels)
#Here we are creating a list of income brackets. We append 250 because we jump from 200 to 250 without 
#including increments of 5 in between. This will be the x-axis of our graph.
```

```python
income_dct = {"Income Level": np.array(income_levels),
                           "Households in Bracket":households}
income_data = pd.DataFrame(income_dct).set_index("Income Level")
income_data.tail()
#We create income_data by converting income_dct to a dataframe. Income_data has columns equal to the keys of
#the original dictionary; 'Income level' and 'Households in bracket', and values equal to an array of the
#income_levels list object and the households object derived from our original income data. Moreover, we
#index the dataframe using the income brackets from the array of income levels.
```

```python
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()
income_data.tail()
#Aggregates number of households by income bracket. Total households in dataset are 128 million, hence, at
#income level 250k+ the households in bracket are 6.375 million while the cumulative number of households
#at or below the 250K+ income bracket is equal to the total number of households in the data; 128 million.
```

```python
income_data.plot(legend=  False)
```

```python
total_households = income_data["Households in Bracket"].sum()
total_households
```

```python
income_data["% in Bracket"] = income_data["Households in Bracket"].div(
    total_households).mul(100)
income_data["Cumulative %"] = income_data["% in Bracket"].cumsum()
income_data
#To find the percent in bracket we divide the households in each bracket by the total number of households 
#in the data and multiply them by 100.
#To find the cumulative % we simply use the dot operator to call the cumsum function of the % in bracket var.
```

```python
plt.rcParams.update({"font.size":16})
fig, ax = plt.subplots(figsize = (15,7))
# create a dataframe that exactly reflects the plot
plot_df = income_data.copy()
plot_df.plot.bar(
    y = ["% in Bracket", "Cumulative %"],
    secondary_y = "Cumulative %",
    fill = True,
    alpha = .75, 
    width = .99,
    legend = True,
    ax = ax)

# add $ in front of each label
x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

# calculate and plot the median value
#  plot_df["Cumulative %"][plot_df["Cumulative %"] >=50]
index = list(plot_df.index).index("60k")
ax.axvline(index + .5, linewidth = 15, color = "C3", alpha = .75,
          label = "Median")
ax.set_title("Median Bracket: \$60k - \$65k\nIncome Distribution in the United States")
#% in bracket on the primary y-axis. cumulative % on the secondary.
```

```python
income_25k_bin = income_data.loc[::5]
income_25k_bin.loc["250k"] = income_data.iloc[-1]
# income_25k_bin["Households in Bracket"] = income_25k_bin["Cumulative"]
# income_25k_bin["Households in Bracket"] = income_25k_bin["Households in Bracket"].diff()
income_25k_bin["Households in Bracket"] = income_25k_bin["Cumulative"].diff()
income_25k_bin["% in Bracket"] = income_25k_bin["Households in Bracket"].div(
    income_25k_bin["Households in Bracket"].sum()).mul(100)
income_25k_bin
#Slice the income brackets by increments of five to create 25k increments
```

```python
plt.rcParams.update({"font.size" : 20})
plot_df = income_25k_bin[["% in Bracket", "Cumulative %"]].iloc[1:]
fig, ax = plt.subplots(figsize = (15, 10))
plot_df.plot.bar(secondary_y = "Cumulative %",
                 width = .99, 
                 alpha = .75,
                 color = ["C0","C1"],
                 ax = ax,
                legend = True)
# ax.legend(loc = "upper center")
x_ticklabels = ["$" + str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

y1_ticklabels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(y1_ticklabels)
ax.set_title("Income Distribution in the United States")
```
