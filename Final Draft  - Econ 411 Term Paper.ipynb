**The Ethanol Blending Program (EBP)**

    India's demand for fuel ethanol has risen steadily since the enactment of the Ethanol Blending Program (EBP) in 2007. This program's purpose was to set a phase-wise blending mandate for Oil Marketing Companies (OMCs), which stipulated that each liter of gasoline consumed for fuel must contain at least 5% ethanol (Ray et al., 2011). The passage of the EBP reflects the government's desire to reduce national reliance on foreign energy imports, thus, increasing India's energy security.

    Since the EBP's adoption in 2007, the program's expansion has been slowed by other economic sectors' ethanol needs and weak domestic production. Consequently, ethanol-blended gasolineâ€™s market penetration has hovered around 2% on average over the past decade (Aradhey, Amit., 2018). Moreover, despite the difficulties associated with achieving the 5% blend target stipulated in the EBP, the legislation's impacts on fuel ethanol demand have still been substantial. In the decade since the GOI enacted the EBP, the transportation sector is now a significant end-user of ethanol, with consumption levels surpassing those from other more established industries at an unprecedented rate (Aradhey, Amit., 2018).

    Furthermore, the gradual expansion of the EBP over the past decade has not gone unnoticed by the Government of India (GOI), which revised guidelines from the original blending program in 2018 to help expedite its expansion. These revisions are tailored towards addressing problems with the original program that arguably hindered its success (Aradhey, Amit., 2018).

    This revised biofuel policy emphasizes the development of the country's domestic ethanol industry and promotes self-sufficiency in ethanol production because India's lack of capacity in ethanol production was arguably responsible for the inability of the EBP to achieve the target blend rates that were initially outlined by the policy in 2007 (Ray et al., 2011)

    Thus, several of the revisions reflect steps toward self-reliance in ethanol utilization for fuel. For example, the EBP's revisions stipulate that ethanol made domestically is rationed for the transportation sector as a priority; thus, other industries' ability to procure ethanol from the domestic market as needed becomes hindered. Consequently, other areas of India's economy that use ethanol often face shortages, which is offset using imports from other countries (Aradhey, Amit., 2018).

    In addition to these changes, the new legislation also calls for alterations to India's trade policy regarding ethanol's purchase from other countries. The revised policy stipulates a ban on the importation of ethanol for fuel blending purposes. Instead, imported ethanol is used explicitly for backfilling supply gaps in the industrial chemicals sector. Moreover, further guidelines are included in the legislation that fixes the price of ethanol for the OMCs responsible for blending gasoline with ethanol at a rate that either meets or exceeds government standards (Aradhey, Amit., 2019).

**Ethanol Demand in India**    

    India's transportation sector is one of its fastest-growing industries and its largest ethanol end-user (Aradhey, Amit., 2019). The steady growth of India's economy and the government's blend requirements have increased the demand for fuel ethanol at an exponential rate. In 2009, the transportation sector consumed 50 million liters of ethanol (Aradhey, Amit., 2019). By 2019, fuel ethanol consumption reached 2.4 billion liters, with the average ethanol content of gasoline reaching a record high of 5.7%, making 2019 the first year that the market penetration of ethanol exceeded the EBP's 5% target (Aradhey, Amit., 2019). 

    Moreover, only two factors have slowed the growth in demand for fuel ethanol. The first has been the preparedness of India's automobile industry in meeting the requirements of the EBP, and the second is demand considerations from other sectors of the economy (Ray et al., 2011). In tests, researchers concluded that the majority of India's vehicular fleet was compatible with 5% ethanol-blended gasoline without the need for significant engine modifications (Ray et al., 2011).

    However, blend rates above 15% will require changes to many of India's existing vehicles (Ray et al., 2011). This barrier has prompted the GOI to work more closely with the nation's leading automobile producers to make the modifications that will make the vehicular fleet compatible with the biofuels policy outlined in 2018 (Aradhey, Amit, 2019)
    

    The use of ethanol as a motor fuel additive is a recent development in India, and as few as three years ago, the demand for fuel ethanol was only the third-largest end-use of ethanol in India (Ray et al., 2011). Currently, around 50% of ethanol demand in India is derived from its use as fuel by the transportation sector. Conversely, the remaining portion of India's ethanol demand is attributed to purposes that are unrelated to its use as a motor fuel additive (Aradhey, Amit, 2019).

    In the past, most of the alcohol produced in India was formed as a byproduct of fermenting sugarcane into alcoholic spirits. The remaining alcohol was used for producing industrial chemicals (Ray et al., 2011). The beverage sector distills sugarcane molasses into undenatured ethanol and combines the ethanol with different flavoring agents to produce alcoholic spirits. The chemical sector typically combines alcohol with a denaturant and uses the denatured alcohol to produce chemicals such as solvents (Ray et al., 2011). These uses for alcohol account for a significant portion of India's total demand.

    The transportation sector has become the point of focus in understanding trends in Indiaâ€™s aggregate ethanol demand because of how rapidly it has overtaken other industries in terms of raw consumption. (Aradhey, Amit., 2019). Moreover, all three sectors are expected to remain significant end users of ethanol, however, the significance of the chemical and beverage sectors are expected to continue shrinking in terms of relative significance to the transportation sector based on assessments of future trends in ethanol demand. (Aradhey, Amit., 2019).

    Moreover, the GOI's steady push to divert ethanol from the chemical industry for use as a motor fuel additive has seen the annual demand consistently outstrip production. The resulting shortfalls in supply for the chemical sector are currently compensated for using imports from foreign countries. This practice has become a consistent theme over the past ten years, with India regularly being a net importer of ethanol because of regular supply shortages  (Aradhey, Amit., 2016). **India's Domestic Ethanol Industry**

    Most of the ethanol India produces is made from sugarcane molasses, a byproduct of sugar production. Three main factors determine sugar production: the area under sugarcane cultivation in hectares, the output of sugarcane per hectare, and the proportion of sugarcane being crushed by sugar mills (Ray et al., 2011). The sugar production level corresponds directly to India's ethanol output because the molasses byproduct needed to produce ethanol is formed only when sugarcane is crushed into cane sugar by a mill (Ray et al., 2011).

    Moreover, the cultivation of Sugarcane in India follows a cyclical pattern. This pattern is characterized by recurring trends, where sugarcane production is high for a 2 to 3-year period and then falls for two years (Ray et al., 2011). The proportion of land allocated for sugarcane cultivation by each farm tends to increase when farmers can sell their crops to sugar mills for higher profits when sugar prices are high for the first few years of the cycle.

    Furthermore, in the years that follow this price decrease, Indian farmers will adjust the mix of crops that they plant because the fall in sugar prices causes a corresponding reduction in the profitability of farming Sugarcane (Ray et al., 2011). Consequently, the proportion of arable land being used to cultivate sugarcane decreases in favor of other profitable crops. This phenomenon indicates a mismatch in the incentive structure for Indian farmers, which culminates in recurring shortages (Ray et al., 2011). The figure below shows the cyclical trend in sugarcane cultivation rates from 2000-2023.
    

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x = data['Year '], y = data["Cane Area (000' Hectares)"], data = data)
ax.set(xlabel = 'Year', ylabel = 'Area''\n'"(000' Hectares)", title = 'Sugarcane Cultivation') 
ax.set_xticks([i for i in range (2000,2024)])
plt.xticks(rotation = 45)
line = ax.lines[0]
line.set_marker('o')
line.set_markersize(10)
plt.show()
```

    This mismatched incentive structure drives the cyclical patterns in sugarcane production that result in a structural problem for the domestic ethanol industry due to industry reliance on molasses, which is never consistent in its availability (Ray et al., 2011). In short, the bumper crop years of the cycle cause ethanol output to increase, which leads to a reduction of India's ethanol deficit and a subsequent decrease in the country's demand for ethanol imports. 

    Conversely, during the sugarcane shortage that follows this initial glut in the cycle, ethanol output falls, causing a corresponding increase in the demand for imports (Ray et al., 2011). Moreover, the demand for sugar also influences ethanol output. When the need for sugar is high, the proportion of sugar crushed in mills increases to meet the demand; this leads to a rise in the availability of byproducts like molasses and a corresponding increase in ethanol production (Aradhey, Amit., 2014). However, when the need for sugar is low, it decreases ethanol output because sugar mills reduce operational capacity, producing less molasses (Ray et al., 2011).
    
    Additionally, the government uses fair and remunerative sugarcane prices to give farmers more certainty they will receive 'fair' prices for their crops each year (Alexander, Mino., 2010).  These â€˜fair pricesâ€™ seek to help farmers by stabilizing the incentive structure of sugarcane farming by insulating farms against the business risk inherent to the lengthy process of cultivating sugarcane instead of other crops (Alexander, Mino., 2010)
         
    However, these price controls may be doing more harm than good in some cases. For example, the government sets remunerative cane prices based on historic farm cost data (Ray et al., 2011). This approach arguably demonstrates the classical downfall of central planning which the GOIâ€™s price controls arguably reflect because in certain years the â€˜fairâ€™ price was arguably set too low causing some farms to underperform because they received insufficient revenue to cover cultivation costs (Alexander, Mino., 2010).

    Conversely, in other years the â€˜fairâ€™ price was arguably set too high which possibly made it difficult for sugar mills to afford to buy sugarcane from farmers to produce sugar. Consequently, mills may have been forced to take on increasing amounts of debt each year to cover their sugarcane arrears to farmers and it could be argued that this debt burden on sugar mills incentives the purchase of less sugarcane from farmers during these years (Ray et al., 2011). 

    Additionally, it could be argued based on these hypothetical arguments against price controls that they should not be used to address any structural problems that exist in Indian agriculture because they may be unintentionally causing some of the observed inconsistencies in sugarcane output to occur each year. 


    However, the use of price controls by the government may become clearer when considerations are made for Indiaâ€™s demographics which may help shed light on the priorities of the GOI and provide arguments that support price controls on sugarcane. Hence, India is the second-most populous country on earth and sugar derived from sugarcane is a major food source in India that the GOI closely controls because food shortages in India have been very costly in the past (Alexander, Mino., 2010). 
    
    Therefore, it can be argued that the â€˜fair and remunerative price for sugarcane reflects the GOIâ€™s desire to control and monitor sugarcane cultivation each year to prevent severe shortages associated with market failure that occurs when prices fail to freely adjust without intervention which arguably occurred in the past (Ray et al., 2011).


    For example, when considerations are made for the versatility of sugarcane e.g., it can be used to produce both food and fuel, and further considerations are made for demographic issues faced by India i.e., food availability issues cause people to go hungry if left unchecked, the GOIâ€™s choice of using price controls becomes a classical tradeoff that lends clarity to an understanding of the structural problem faced by Indiaâ€™s domestic ethanol industry (Gulati, A., and T. Kelley, 1999).

    Theoretically, India could produce all the ethanol it needs. Though, it would have to augment its production of ethanol using sugarcane itself, food grains, and sugarcane juice in conjunction with the molasses already being utilized (Aradhey, Amit., 2018). However, this approach forces a choice between using sugarcane and food grains for food or fuel, which describes a difficult trade-off the GOI is committed to avoiding. This sentiment is evidenced by legislative guidelines introduced in 2015, which prohibit the use of raw sugarcane as a feedstock to produce ethanol (Aradhey, Amit., 2019). 

    Hence, sugarcane molasses remains the primary feedstock for ethanol production in India. However, alternative feedstocks for ethanol are being explored but require further development before they can be considered viable alternatives to molasses (Aradhey, Amit., 2019).

    Moreover, resolving the structural problems that pervade India's ethanol market is of great interest to the government per its rhetoric; however, discovering a pragmatic solution remains elusive as other considerations take precedence over the resolution of any structural problems faced by the domestic market which arguably reflects good priorities clashing with good intentions (Ray et al., 2011). Thus, ethanol production will remain inconsistent so long as sugarcane molasses remains the primary feedstock in its production (Ray et al., 2011).

**Patterns in Trade: Ethanol Imports and Exports**


    The policy guidelines of India's EBP do not allow the importation of ethanol for fuel blending purposes. Moreover, the GOI has imposed a 150% import duty on all imports of undenatured ethanol coming into the country (Aradhey, Amit., 2018). Thus, a combination of import restrictions and trade barriers have led to nearly all the ethanol that India imports each year being denatured (Aradhey, Amit., 2019).

    Furthermore, these imports of denatured ethanol are used to backfill any supply deficits for the chemical sector when ethanol rationing for use as motor fuel renders the domestic supply inadequate. Hence, when ethanol demand outstrips domestic production during a poor cane harvest, India's ethanol deficit grows. Conversely, when ethanol production is strong due to a sugarcane bumper harvest, the ethanol deficit shrinks (Aradhey, Amit., 2018). 

    The figure below illustrates trends in import levels from 2000-2023

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x = data['Year '], y = data["Imports Ethanol (ML)"], data = data)
ax.set(xlabel = 'Year', ylabel = 'Volume''\n'"(ML)", title = 'Ethanol Imports') 
ax.set_xticks([i for i in range (2000,2024)])
plt.xticks(rotation = 45)
line = ax.lines[0]
line.set_marker('o')
line.set_markersize(10)
plt.show()
```

    Moreover, the supply deficit for 2020 is expected to narrow, given steadier levels of domestic production. However, imports will continue to be used to augment supply as needed. For instance, India's import demand of 716 million liters in 2017 was the highest year on record for import quantities (Aradhey, Amit., 2018). Thus, it can be concluded that India's capacity to produce ethanol is improving. However, the volumes of ethanol Imported in recent years indicate that self-sufficiency has yet to be achieved.

    Furthermore, most of the ethanol that India imports comes from the United States. In the past six years, the share of ethanol imports represented by the U.S. grew by nearly 22%, and by 2017 U.S. ethanol represented 96% of all ethanol imported by India. The remaining 4% was imported from China, Pakistan, South Korea, and Bhutan (Aradhey, Amit., 2018).
     
    In contrast, patterns in export levels are not as straightforward in terms of intuition as those observed in imports. However, the mystery of why export levels are often high when the country runs a domestic deficit unravels when a distinction between the two types of ethanol is considered. For instance, virtually all the ethanol that India exports each year is undenatured, while its imports consist exclusively of denatured ethanol (Aradhey, Amit., 2018). 


    These exports are produced by distilleries using molasses from cane to make alcoholic spirits profitable enough to sell in global markets to warrant importing denatured ethanol from the U.S. to backfill the resulting supply gap in domestic markets (Ray et al., 2011).

**Market Dynamics: The Sugar Cycle**

    In this section, a hypothetical analysis of sugarcane price movements under the condition of free competition is explored and the rationale behind the GOIâ€™s â€˜fairâ€™ cane price is considered. This section contains speculation by the author for illustrative purposes that may contribute to a more complete understanding of the issues explored in this research without losing sight of its objectives.

    Hence, sugarcane is theorized to mirror the sugar cycle in price and quantities, with adjustments in the cane cycle lagging those in the sugar cycle by around one year on average. This lag is attributed to the 16-month process of cultivating sugarcane, which delays cane price adjustments until output quantities are realized the year following (Ray et al., 2011).

    Furthermore, the cane price is theorized to move opposite sugar prices because increased sugar production drives sugar prices down while simultaneously increasing the price mills are willing to pay for the cane to meet rising input demands. Conversely, decreased sugar production causes sugar prices to rise while driving cane prices down as input demand falls (Ray et al., 2011). 

    Moreover, this lag in the adjustment of cane quantity to those occurring in sugar quantity each year may have some interesting ramifications for Indian farmers. For example, this lag characterizes a structural problem in Indian agriculture which is derived from the inability of sugarcane prices to adjust to those occurring in sugar at rates quickly enough to prevent inefficient outcomes (Ray et al., 2011). In other words, the cane price is unable to change rapidly enough in response to those occurring in sugar to clear the cane market fully each year.

    Consequently, these slow adjustments in cane prices arguably cause asymmetries in farmers' information as they decide what crops to cultivate during their planting season. This lagged adjustment in cane prices arguably makes cane growing a risky business for the farmers choosing to plant cane each year with each decision regarding how much to produce carrying risks that could bankrupt a farm if the wrong gamble is made in some cases. 

    Therefore, farmers actively rely on cane prices to guide their decisions for what mix of crops to plant today, which is variable through time in response to previous sugar price movements. Hence, India's farmers are left with no choice but to sell their crops at unfavorable prices when movements in price are difficult to anticipate. (Ray et al., 2011). This dynamic indicates a lead-lag effect where farmers use rational expectations regarding future price trends in sugar to help them anticipate its movements to stay 'ahead of the curve' as cane price lags those occurring in sugar today.

    Hence, the cultivation of cane in India is arguably not unlike the futureâ€™s market in American agriculture. For example, under the hypothetical assumption of free competition, the Indian farmers would arguably use speculative guesswork to anticipate what direction sugar and cane prices will move in, which is arguably like speculation by U.S. farmers about trends in U.S. commodity prices that may lend an edge as they vie to 'beat the market'. However, India's cane market departs its similarities from markets like those from the U.S. with the government's price controls on sugarcane (Ray et al., 2011).

    The government's use of â€˜fairâ€™ cane prices could arguably be called an attempt to address the structural problems inherent to Indian sugarcane farming using government intervention and central planning to address a form of market inefficiency that is arguably a legitimate example of market failure that threatens negative outcomes if left unchecked. 

    Conversely, these price floors which arguably reflect a well-intentioned attempt to clear an inefficient market by ensuring farmers always get 'fair' prices for their crops and sugar is always available for Indiaâ€™s people may have unintended consequences that are noteworthy for the sake of nuance in research. For example, this policy of using price controls may be harmful to farmers as the remunerative cane price is sometimes tricky for sugar mills to afford, forcing them to take on increasing amounts of debt to cover their sugarcane arrears owed to farmers each year (Ray et al., 2011). 

    Consequently, interest rates play an essential role in a mill's decisions about how much sugar should be produced in a year as the opportunity cost of borrowing the money needed to ensure its operation is paramount to its profitability. Moreover, the GOI also redistributes around ten percent of all sugar produced by sugar mills in the calendar year to needy families. This allocation is known as levy sugar, which the government procures from sugar mills at cost using a quota basis (Alexander, Mino., 2010).

    Hence, these dynamics manifest themselves as the price movements behind each of the interrelated commodity markets, which make up a more significant sugar cycle. Moreover, the adjustment mechanisms observed for sugarcane prices are thought to be responsible for the structural flaws that have perpetuated the 'sugar cycles' apparent within India's agriculture system that has proven challenging to address pragmatically even with price controls from the GOI.

**Market Dynamics: Prices and Land Use**

    India's farmers use 2-3-year crop rotation systems based on nutrient considerations, cultivation times, and the relative price and marginal costs of cultivation associated with each crop. The farmer selects which crops to include in their rotation based on the relative profitability of crops which are substitutes in farming, meaning one cannot be grown consecutively after the other on the same land; they must be produced simultaneously in fields with heterogenous crop mixes or separately in areas used for cultivating one unique crop at a time (Gulati, A., and T. Kelley, 1999).

    Furthermore, considerations are made for the relative profitability of complementary crops, meaning they should be grown in consecutive order on the same land because complementary crops provide some of the nutrients needed by each other when cultivated in a specific order on the same field. These fields are known as 'twice-sown,' meaning they are used to grow two or more crops in the same year (Gulati, A., and T. Kelley, 1999).

    These 'twice-sown' fields are the lynchpin of the crop rotations that occur in India, with their cultivation involving a trade-off in the amount of arable land allocated to a given crop within some set of substitute crops. This trade-off assumes that substitute crops must either be jointly cultivated in the same crop rotation cycle or cultivated in an entirely different and unrelated crop rotation cycle (Gulati, A., and T. Kelley, 1999).

    Therefore, the cultivation of any substitute crops within the same crop rotation sequence must be grown jointly on the fixed amount of arable land available for agriculture at that time. This produces the trade-off in land use mentioned earlier because the available land is mostly fixed through time. However, the number of ways this land can cultivate different crops is virtually unlimited (Gulati, A., and T. Kelley, 1999).

    Moreover, this trade-off in allocating arable land between substitute crops being cultivated jointly can be conceptualized as a theoretical framework that treats crop rotation patterns as a response by farmers to changing economic conditions to optimize their profits. Moreover, crop rotation changes occur as farmers' respond' to dynamic changes occurring in the relative profitability of different crop varieties, hence, adjusting the mix of crops planted accordingly.

    Furthermore, the substitution of varying crop varieties based on their profitability occurs within a subset of crops similar to their nutrient needs, cultivation length, and soil requirements. Moreover, each variety's relative profitability is gauged by the farmer using current market prices, price expectations for the future, and present cultivation costs (Gulati, A., and T. Kelley, 1999).

    Subsequently, certain assumptions are imposed on this research's theoretical framework to explain patterns in crop rotation. These assumptions allow for variations in land use to occur in terms of adjustments to all crops' rotation within a variety based on the shifts in profitability occurring between varieties over time.

    Hence, the theoretical framework assumes that each successive set of crops cultivated jointly during each cycle is considered strict substitutes in cultivation, starting from the initial set ending with the final set. Moreover, each subsequent group of substitute crops cultivated per the pattern specified for crop rotation is assumed by the model to complement each other in cultivation (Gulati, A., and T. Kelley, 1999).

    Moreover, the set of complementary crops are grown in separate periods based on temporal considerations for the overall length of each crop rotation cycle and the nutrient considerations of each crop that will be developed during the cycle, which affects the order of cultivation in complementary crops through time (Gulati, A., and T. Kelley, 1999). 

    Furthermore, these assumptions allow the theoretical model to ensure that the estimation of a crop rotation cycle conforms to proper crop rotation practices, e.g., an actual farmer would not plant substitute crops consecutively in the same 'twice-sown' field because it would reduce yields and by extension profits, hence, the importance of these assumptions (Gulati, A., and T. Kelley, 1999).

    For example, cane cultivation requires 16-months to fully cultivate while rice takes around 8-months and wheat around 4-months. Moreover, rice and wheat are complements in farming to each other and substitutes in cultivation for Sugarcane (Gulati, A., and T. Kelley, 1999). Thus, a farmer can choose to set aside a plot of land to plant a homogenous field of sugarcane, which requires 16-months to cultivate fully, or this same plot of land can be used to plant a 'twice-sown' field containing both wheat and rice grown consecutively in a rotation that begins and ends with wheat with rice being cultivated in the middle of the cycle.

    Therefore, the total length of time required by a farm to grow wheat and rice in the order above is around 16-months, and sugarcane itself takes 16-months to cultivate fully. Hence, the farmer faces a clear trade-off between growing Sugarcane or growing wheat and rice (Gulati, A., and T. Kelley, 1999). Thus, the ebb and flow like patterns in sugarcane output that occur each year across all of India are a function of profitability associated with cane cultivation relative to the profits of planting substitute crops for sugarcane, which are complementary in their agriculture, allowing them to be grown using crop rotation, e.g., rice and wheat (Alexander, Mino., 2010).

**The Empirical Model**

    A simultaneous equation model is used to estimate each commodity's output quantities that make up a broader sugar cycle component. This is done by treating commodity prices as endogenous, allowing the cane output pattern to be estimated based on developments in the sugar market, ultimately thought to drive the sugarcane market. 

    Furthermore, the cycles in cane output lag those occurring in sugar due to its 16-month cultivation cycle delay price adjustments for sugar cane that move opposite those occurring in sugar (Gulati, A., and T. Kelley, 1999). Hence, cane farmers' cultivation choices are made each year using uncertain and incomplete information, which causes cane output to follow predictable patterns of contraction and expansion (Ray et al., 2011).

    Moreover, sugar production has a near-perfect one-to-one relationship with molasses output and ethanol production through its effect on cane cultivation rates. The quantity of cane is ultimately used to produce sugar by mills. The lag in cane cultivation causes the adjustment in cane prices to fall behind those occurring in sugar (Ray et al., 2011). 

    Thus, years that experience a cane shortage that slashes both ethanol and molasses production levels tend to happen when a lag in cane price from the previous year signaled farmers to scale production down when production should have been increased. Conversely, some years see an opposite effect where the lag in the adjustment of cane price sends the wrong signal to farmers that output should be increased in the current year when it should be scaled back, causing an oversaturation of the cane market the following year, which tends to increase the supply of ethanol and molasses (Alexander, Mino., 2010).

    Furthermore, the sugar production trend affects both ethanol and molasses output but in real-time because any expansion or contraction in sugar manufacture impacts the proportion of cane being crushed into sugar by mills. This is significant because the molasses feedstock needed by India's domestic ethanol industry to produce its product is formed only when sugarcane is crushed into sugar-producing the needed molasses as a byproduct (Ray et al., 2011).

    Furthermore, India's domestic supply of ethanol is then differenced from domestic demand levels to determine if supply shortages or surpluses are evident each year. The internal balance can predict the country's demand for imported ethanol with an expectation that import quantities will rise in years of ethanol shortage in the local market. Conversely, ethanol export levels are predicted to shrink with shortages and grow with surpluses. However, export quantities are treated as exogenous for simplicity. 

    Lastly, a trade identity is specified, constraining the entire system to ensure the full ethanol availability and the aggregate ethanol demand in each year (t)  are equivalent. This ties the model together by serving as the market-clearing identity that relates the otherwise independent sub-systems to each other using the ending stocks as the left-hand side variable because it is known with the least certainty.

    In this section, we diverge from theory to mitigate issues with data availability. We will take sugarcane prices as given because they are set by the government. Also, we will estimate behavioral equations on an individual basis using OLS due to a lack of observations overall and a lack of observations for ethanol prices. 

    Furthermore, we will make assumptions about India's sugar market and the relationship it has with the ethanol market. Firstly, we will describe India's demand for sugar using consumer incomes, sugar prices, and gur prices. Next, we will specify an identity that assumes the sugar market clears at prevailing price levels and we will create a rough estimate of sugar production using sugar demand, exports of sugar, and imports of sugar. For simplicity, we take import and export levels in the sugar market as given.

    Hence, our model for sugar production is based on an estimation of sugar demand factoring for variances in the global market by taking sugar import and exports levels as given. Next we will make assumptions about the relationship between the ethanol market and the sugar market. We will use the average proportion of cane allocated to sugar production and our estimate of sugar production to estimate the amount of cane being used to produce sugar in a given year. Next, we will use this estimate for cane allocated to sugar production to produce an estimate for the amount of sugarcane mollasses produced each year. we will do this by creating an identity that multiplies the amount of cane being crushed into sugar by the average proportion of sugarcane mollasses produced from this process. Lastly, we will assume that India ferments all of this mollasses into ethanol at a constant rate. Hence, using these identites we can create an estimate for India's production of ethanol each year using our estimate for sugar production.

    In this section we will run three regressions to estimate the relationships between consumer incomes, prices, and sugar demand. The first regression will estimate these relationships using log-transformed annual data from 2000-2023. We will discuss the statistical significance of the results and any econometric problems the model has. Moreover, we expand upon this model to help mitigate any problems that we observe.

```python
model_vars = ['Year ', 'Sugar $ (USD/MT)', 'Gur $ (USD/MT)', "India's GDPPC (USD)", "Demand Sugar (000' MT)"]
model_data = data[model_vars]
model_data = pd.DataFrame(model_data)
model_data
```

```python
reg_data = model_data.apply(pd.to_numeric, errors = 'coerce')
```

```python
log_reg_data = reg_data.iloc[:,1:].apply(np.log)
log_reg_data
```

```python
x_vars = ["Sugar $ (USD/MT)", 'Gur $ (USD/MT)', "India's GDPPC (USD)"]
y_var = ["Demand Sugar (000' MT)"]
```

```python
import statsmodels.api as sm
y = log_reg_data[y_var]
X = log_reg_data[x_vars]
X["Constant"] = 1
results = sm.OLS(y, X).fit()
```

```python
results.summary()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data_with_const = sm.add_constant(log_reg_data)
vif_data = pd.DataFrame()
vif_data["feature"] = data_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(data_with_const.values, i) for i in range(data_with_const.shape[1])]
print(vif_data)
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
for column in log_reg_data.columns:
    result = adfuller(log_reg_data[column])
    print(f'ADF Statistic for {column}: {result[0]}')
    print(f'p-value for {column}: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
    print('\n')
```

    In this first regression, we get a price elasticity of demand for sugar of -0.3831. This indicates that a 1% increase in sugar prices will lead to a -0.3831 decrease in the demand for sugar. Moreover, the t-statistic for sugar prices is -2.467 indicating that sugar prices are statistically significant at the 95% confidence interval. The coefficient for gur prices is 0.3097. This indicates that a 1% increase in gur prices will lead to a 0.3097% increase in the demand for sugar. Furthermore, the t-statistic for gur prices is 2.230 indicating that it is statistically significant at 95% C.I. Both coefficients for prices have the correct sign, and their respective coefficients fall within reasonable ranges. The coefficient for GDPPC is 0.3309 meaning that a 1% increase in consumer incomes leads to a 0.3309% increase in the demand for sugar. This coefficient has a t-stat of 9.105 indicating statistical significance at the 95% C.I. and the sign and the magnitude of the coefficient support the theoretical basis of the model.

    This first regression was not de-trended to account for non-stationarity in time-series data. The Augmented Dickey-Fuller test indicates that sugar demand has a unit root and is non-stationary. Therefore, it follows that further refinement of this model is needed. Moreover, the variance influence factors (VIF) for each variable are high. This indicates that multicollinearity is an issue in this model as well.

    In the second regression, we will de-trend the data by taking the first difference of each variable. Data involving prices and incomes often exhibits a trend due to inflation and real growth over time. Hence, these refinements may improve our results. 

```python
alternative_model_data_1 = log_reg_data[['Sugar $ (USD/MT)', 'Gur $ (USD/MT)', "India's GDPPC (USD)","Demand Sugar (000' MT)"]].diff()
alternative_model_data_1
```

```python
alternative_model_data_1 = alternative_model_data_1.dropna()
alternative_model_data_1
```

```python
for column in alternative_model_data_1.columns:
    result = adfuller(alternative_model_data_1[column])
    print(f'ADF Statistic for {column}: {result[0]}')
    print(f'p-value for {column}: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
    print('\n')
```

```python
y = alternative_model_data_1[y_var]
X = alternative_model_data_1[x_vars]
X["Constant"] = 1
results = sm.OLS(y, X).fit()
results.summary()
```

```python
data_with_const = sm.add_constant(alternative_model_data_1)
vif_data = pd.DataFrame()
vif_data["feature"] = data_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(data_with_const.values, i) for i in range(data_with_const.shape[1])]
print(vif_data)
```

```python
fig, ax = plt.subplots(figsize = (12,8))
results.resid.plot.hist(bins = 15, ax=ax)
```

    In this second regression, we refined the model by de-trending the data. The ADF test indicates that both sugar prices and sugar demand still have problems with non-stationarity, but the problem is not as severe. Moreover, the VIF test indicates that multicollinearity is still an issue, but it is not as problematic. Furthermore, the t-stats,p-values, and coefficients have the correct sign and magnitude except for the constant which is now statistically insignificant. I theorize the constant was capturing the effects of the lag of sugar demand on its present level value, and upon de-trending the data the value of the constant is now negligible.  

    In our third regression, we attempt to incorporate elements of vector autoregression using the lagged percent change in sugar price as an independent variable to capture potential seasonal effects in our model. Moreover, the DW test statistic indicates autocorrelation of the error terms in the second regression. I theorize that past economic conditions and seasonality may cause this correlation. Hence, we refine the model further at the risk of over-fitting.

**Conclusions**

    This paper provides a brief overview of the relationship between the ethanol and sugar market in India. It attempts to model the country's import demand for ethanol as a function of domestic deficits in ethanol production related to structural problems in the sugar market and commodity prices within the broader ethanol market. A combination of limited data availability and market complexity make India's ethanol market difficult to model.

    Moreover, as the EBP continues to increase in scope it can be expected that India's demand for ethanol will continue to increase at a very high rate. However, India's capacity for producing ethanol is also increasing rapidly. The use of feedstocks like B-heavy molasses and sugarcane juice to produce ethanol greatly increases the amount of feedstocks that can be used for ethanol fermentation. Hence, the domestic industry is not as under-developed as it once was.

    A more complete understanding of the market would need to build on some of the ideas that we analyzed in this paper. It is not just the sugar cycle and molasses production that drives India's production of ethanol anymore. The country now has more ethanol refineries and more input options for ethanol production than it has in the past. These new dynamics would need to be studied closely to create a more complete understanding of the market. Additionally, price data on ethanol would need to be acquired to create an empirical model that represents the market more accurately.
