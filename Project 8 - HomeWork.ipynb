**.1**

Generate data for one random variable with a mean of 5 and standard deviation of 1. Run a one-sample t-test that the mean of the sample is different than 0. Perform the same test again with a mean of 1 and a standard deviation of 1.

```python
import numpy as np
from scipy import stats
mean_5 = 5
std_dev_1 = 1
random_var = np.random.normal(mean_5,std_dev_1, 100)
```

```python
t_statistic_5, p_value_5 = stats.ttest_1samp(random_var, 0)
print("T-statistic for mean 5:", t_statistic_5)
print("P-value for mean 5:", p_value_5)
```

```python
mean_1 = 1
std_dev_1 = 1
random_var1 = np.random.normal(mean_1,std_dev_1, 100)
```

```python
t_statistic_1, p_value_1 = stats.ttest_1samp(random_var1, 0)
print("T-statistic for mean 1:", t_statistic_1)
print("P-value for mean 1:", p_value_1)
```

**.2**

Why has the result changed? In answering this question, plot both samples with the index as the x-axis values. Use two colors to differentiate the samples. Indicate the value of zero using a horizontal line. Reference the plot in your answer.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(range(len(random_var)), random_var, color='blue', label='random_var')
plt.scatter(range(len(random_var1)), random_var1, color='orange', label='random_var1')
plt.axhline(y=0, color='r', linestyle='--', label=0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.title('Random Variable Samples')
plt.show()
```

The two samples have different t-statistic values and different p-values
because the random variables were drawn from a normal distribution with different means. The t-test we conducted tests against a null hypothesis that the mean of each random variable is zero. When we conduct a t-test for the dataset 'random_var' which has 100 observations drawn from a normal distribution with mean = 5 and std.dev = 1 and alpha = 0.05, we get a t-stat of 48.102 and p-value of 1.82e-70. These results indicate that we can be 95% confident that the mean of this dataset is not equal to zero. Conversely, the dataset 'random_var1' was drawn from a normal distribution with a mean of 1. Hence, the t-statistic is far smaller and the p-value far greater for the second dataset because the mean of this dataset is not as far away from the hypothesized value of zero.

Now run a t-test for the two samples. Print the results and explain their meaning.

```python
t_statistic, p_value = stats.ttest_ind(random_var, random_var1)
print("T-statistic for the two samples:", t_statistic)
print("P-value for the two samples:", p_value)
```

Here we test if the sample means for the dataset random_var and random_var1 are different. The null hypothesis that we are testing against states that the difference between the two means is zero. The t-stat of 25.77 and p-value of 3.59e-65 indicate that we can be 95% confident that the difference between the two means is not equal to zero with n = 100 and alpha = 0.05.
