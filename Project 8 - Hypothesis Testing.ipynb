```python
import numpy as np
import pandas as pd
mutual_fund_data = pd.read_csv("Top100MutualFunds.csv")
mutual_fund_data
#I downloaded the "Top100MutualFunds.csv" from your github and put it in
#my working directory.
```

```python
import pandas_datareader.data as web
import yfinance as yf
from datetime import datetime

mutual_fund_dict = {}
symbols = list(mutual_fund_data["Symbol"])
start = datetime(1980,1,1)
end = datetime(2020, 1, 1)
(start, end, symbols)
#The 'symbol' column denotes the acronym for each mutual fund in our data
```

```python
yf.pdr_override()
fund_data = web.get_data_yahoo(symbols, start, end)
fund_data
#The override function is used to override the default pandas data reader
#so we can use yahoo finance data reader to fetch financial data.
```

```python
# .dropna(how = "all", axis = 1) -- remove any column that has only nan observations
close_data = fund_data["Adj Close"].dropna(how = "all", axis=1)
close_data.tail()
#Here we declare the 'close_data' object as the adjusted close data from
#the fund dataset and we drop all columns with non-applicable values
```

```python
close_data["S&P 500"] = web.get_data_yahoo("^GSPC", start, end)["Adj Close"]
#Here we column bind adjusted close data from the S&P 500 to our close_data
#object
```

```python
rolling4wk = np.log(close_data).diff(20)
rolling4wk.tail()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (20,10))
rolling4wk.plot.line(ls = "", marker = ".", ax = ax, color = "k", markersize = 1, alpha = .1, legend = False)
rolling4wk["S&P 500"].plot.line(color = "C3", ax = ax, legend = False)
```

```python
monthly_close_data = close_data.resample("M").first()
monthly_close_data.tail()
#This code resamples the time-series data contained in 'close_data'
#using a monthly interval. The first() function selects the first data
#point within each monthly interval and these values make up the data of
#of the 'monthly_close_data' object. 
```

```python
monthly_close_data.index
#This code returns the index of the monthly close data. Because each the
#data is specified with a monthly delta each year should have 12 obs. However,
#NA values were dropped so each year has 10 observations.
```

```python
monthly_returns = np.log(monthly_close_data).diff()
monthly_returns.tail()
#This code computes the log of the monthly close data and then takes the
#first difference of these values to calculate monthly returns.
```

```python
avg_monthly_return = monthly_returns.loc["2012":].dropna(axis = 1)
avg_monthly_return.mean(axis=1)
#This code selects the rows in the monthly_return dataframe from 2012 onwards
#and drops any columns with an NA value. Finally, the average return for
#each month is calculated using the 'mean' function.
```

Hypothesis Test

One Sample T-test

 $$t = \frac{{\bar{X} - \mu}}{{s/\sqrt{n}}}$$ 

$$H_0: \mu = \mu_0$$

$$H_1: \mu \neq \mu_0$$

$$\mu=0$$

```python
means = monthly_returns.mean()
ses = monthly_returns.sem()
means, ses
#We need means and standard errors from the monthly_returns data to calculate
#t-stats for hypothesis testing
```

```python
one_sample_test = means.div(ses)
#Calculate the t-stats
fig, ax = plt.subplots(figsize = (20,10))
one_sample_test.plot.hist(ax = ax)
ax.axvline(one_sample_test.mean(), color = "k", ls = "--")
#Calculate the mean value of the t-stats and insert a line at that value
#along the x-axis
ax.set_title("$H_0: \\bar{X}=0$")
#Set the title of the diagram to the null hypothesis using LaTex
ax.set_xlabel("t-values")
#Set the x-label
```

Two Sample T-test

$$t = \frac{{\bar{X}_1 - \bar{X}_2}}{{\sqrt{\frac{{s_1^2}}{{n_1}} + \frac{{s_2^2}}{{n_2}}}}}\$$

```python
# s^2 / n
sn = monthly_returns.var().div(monthly_returns.shape[0])
two_sample_test = means.sub(means["S&P 500"]).div(sn.add(sn["S&P 500"]).pow(.5))
two_sample_test
```

```python
fig, ax = plt.subplots(figsize = (20,10))
two_sample_test.plot.hist(ax = ax)
ax.set_title("$H_0: \\bar{X_1}=\\bar{X_2}$")
ax.set_xlabel("t-value")
```
